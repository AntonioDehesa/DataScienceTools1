# Q4(points 1=.5+.5) Using *ast* module  convert entities to python dictionary. This dictionary should contain hashtags and user_mentions
 - Create a columns named hashtags. Values in this columns should be comma separated values of **text** attribute in **hashtags**  value.
 - Create a columns named user_mentions.Values in this columns should be command separated values of **name** attribute in **user_mentions** values.

```python
# write code here
def get_hashtag(main_dict):
    try:
        d = ast.literal_eval(main_dict)
        hashtags = d["hashtags"][0]
        text = hashtags["text"]
    except:
        text = np.nan
    return text

def get_name(main_dict):
    try:
        d = ast.literal_eval(main_dict)
        user_mentions = d["user_mentions"][0]
        name = user_mentions["name"]
    except:
        name = np.nan
    return name

climate_feed_df["hashtags"] = climate_feed_df["entities"].apply(get_hashtag)
climate_feed_df["user_mentions"] = climate_feed_df["entities"].apply(get_name)
climate_feed_df[~climate_feed_df["text"].isna()].sample(100)
```

# Q 6(point 1= .5+.5) Remove all the reference  of RT and  @user: or @user, and url in climate_feed_df text attribute.

```python
# write code here
# Remove https
def remove_noise(orig_tweet):
    x = re.sub(r"https\S+", "", orig_tweet)
    x = re.sub(r"\\n*", "", x)
    x = re.sub(r"\@\S+", "", x)
    x = re.sub(r"RT\S+", "", x)
    # Or a single regex with "https\S+|\\n*|\@\S+|RT\S+"gm, but its the same
    return x
#df["original_tweet_text"] = re.sub("https\S+" ,"",df["original_tweet_text"])
df_valid = climate_feed_df[~climate_feed_df["original_tweet_text"].isna()]
climate_feed_df["original_tweet_text"] = df_valid["original_tweet_text"].apply(remove_noise)
def remove_noise2(orig_tweet):
    x = re.sub(r"https\S+", "", orig_tweet)
    x = re.sub(r"\\n*", "", x)
    x = re.sub(r"\@\S+", "", x)
    x = re.sub(r"RT", "", x) # For some reason, RT from text field does not get removed with the previous regex
    # Or a single regex with "https\S+|\\n*|\@\S+|RT\S+"gm, but its the same
    return x

df_valid = climate_feed_df[~climate_feed_df["text"].isna()]
climate_feed_df["text"] = df_valid["text"].apply(remove_noise2)
climate_feed_df.head()
```

# Q 7 (2 point) Create a new attribute in climate_feed_df named *final_tweet_text*. If *original_tweet_text* is None  then use climate_feed_df.text otherwise use *original_text_tweet* if initial words matched with *climate_feed_df.text* words. If words doesn't match use climate_feed_df.text

```python
# wite code here
climate_feed_df["final_tweet_text"] = ""
index = climate_feed_df.index[climate_feed_df["original_tweet_text"].isna()]
climate_feed_df["final_tweet_text"].iloc[index] = climate_feed_df["text"].iloc[index]
```

```python
df_temp = climate_feed_df[~climate_feed_df["original_tweet_text"].isna()]
def filt(row):
    try:
        orig_tweet = row.original_tweet_text.lower()
        orig_tweet = re.findall(r'\w+', orig_tweet)[:3]
        text = row.text.lower()
        text = re.findall(r'\w+', text)[:3]
        return text == orig_tweet
    except:
        return False # return false, as text is NAN
# Index 2 = elements which first 3 words in original_tweet_text match first 3 words in text
index2 = df_temp.index[df_temp.apply(filt,axis=1)]
# Index 3 = elements whose original_tweet_text is not NAN, and the first 3 words do not match the first 3 words in text
index3 = df_temp.index[~df_temp.index.isin(index2) & ~df_temp.index.isin(index)]
climate_feed_df["final_tweet_text"].iloc[index2] = climate_feed_df["original_tweet_text"].iloc[index2]
climate_feed_df["final_tweet_text"].iloc[index3] = climate_feed_df["text"].iloc[index3]
climate_feed_df[~climate_feed_df["final_tweet_text"].isna()]["final_tweet_text"]
```
