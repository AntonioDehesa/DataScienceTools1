# Assignment 5, Part 2:  Make a list of strings for nine Greek letters, ‘alpha’, for example.
Make that list such that they are not in alphabetic order.  Make two 9-element numpy arrays
of random floating-point numbers with the estimated mean 10 and standard deviation 1.5.
Make another array of nine elements ranging from zero to two times pi.  Name it ‘angle’.
Make another array holding the cosine of that ‘angle’ array.  Construct a dictionary from all
of the above.  Form a DataFrame from that dictionary and print it out.  Sort the DataFrame
ascending on the Greek letters, drop two columns of your choice, drop one of the rows, and
print that out.

```python
greek_letters = ["alpha", "beta", "gamma", "delta", "epsilon", "zeta", "eta", "theta", "iota"]
shuffle(greek_letters)
n_arr1 = np.random.normal(10, 1.5, 9)
n_arr2 = np.random.normal(10, 1.5, 9)
angle = np.random.uniform(0, 2*pi, 9)
cos_angle = np.cos(angle)
my_dict = {}
for i in range(len(greek_letters)):
    my_dict[greek_letters[i]] = (n_arr1[i], n_arr2[i], angle[i], cos_angle[i])
df = pd.DataFrame.from_dict(my_dict, orient="index")
df.columns = ["n_arr1", "n_arr2", "angle", "cos(angle)"]
df
```

```python
df = df.sort_index(ascending=True)
df = df.drop(["n_arr1", "cos(angle)"], axis=1)
df = df.drop(["zeta"])
df
```

# Assignment 5, Part 3:  Write a program in Python to create and print out the first twelve Fibonacci numbers.
Then iterate over the last five numbers to build another list with the ratio of each number to its predecessor.
What do you observe about this latter list?

```python
fibres = [0, 1]
def fibo(n):
    if n < 0:
        print("Invalid input")
    elif n < len(fibres):
        return fibres[n]
    else:        
        fibres.append(fibo(n - 1) + fibo(n - 2))
        return fibres[n]
fibo(12) # This is just to build the array up to the 12th number
print(fibres)
last_five = fibres[8:]
predec = fibres[7:len(fibres)-1]
ratios = [m/n for m,n in zip(last_five, predec)]
print(ratios)
# We can see that the ratios are approximately the golden ratio
```

# Assignment 5, Part 4:  
Provide a function that converts temperature in Kelvin to Rankine.  Make a list of
five Kelvin temperatures and print out their values in Rankine.  Repeat using a lambda function.

```python
# R = K*9/5
# k = 5R/9
def kelvin_to_rankine(k):
    return k*9/5
kelvin_temps = [0,150, 373.15, 273.15, 951]
rankine_temps = [kelvin_to_rankine(x) for x in kelvin_temps]
print(rankine_temps)
```

```python
kel_to_ran = lambda x: x*9/5
new_res = [kel_to_ran(x) for x in kelvin_temps]
new_res
```

# Assignment 5, Part 5:  
Construct a Pandas dataframe with two columns of 100 random numbers drawn from a normal
(Gaussian) distribution.  Divide the DF into two with a 75-25 split at random.  Reset the indices of each new DF.
Print the first 12 lines of each.  Compute the mean and standard deviation of each column of each new DF.

```python
df = pd.DataFrame(data=np.array([np.random.normal(0,0.1,100),
                                 np.random.normal(0,0.1,100)]).reshape(100,2), 
                  columns=["Col1", "Col2"]) # Create the dataframe with two columns of 100 elements with a normal distribution
df
```

```python
df_75 = df.sample(frac=0.75)
df_25 = df.drop(df_75.index) # Splits the dataframe into two: one with 75%, and the rest in another df
df_75
```

```python
df_25
```

```python
df_75 = df_75.reset_index(drop=True)
df_25 = df_25.reset_index(drop=True) # Resets the index
# Prints the first 12 elements of each df
print(df_75.head(n=12))
print(df_25.head(n=12))
```

```python
print("DF_75")
print("Col 1")
print("Mean")
print(mean(df_75["Col1"]))
print("Standard deviation")
print(stdev(df_75["Col1"]))
print("Col 2")
print("Mean")
print(mean(df_75["Col2"]))
print("Standard deviation")
print(stdev(df_75["Col2"]))

print("DF_25")
print("Col 1")
print("Mean")
print(mean(df_25["Col1"]))
print("Standard deviation")
print(stdev(df_25["Col1"]))
print("Col 1")
print("Mean")
print(mean(df_25["Col2"]))
print("Standard deviation")
print(stdev(df_25["Col2"]))
```

# Assignment 5, Part 6:  
Load the Titanic dataset (a Seaborn built-in) into a DF.  Compute the survival rate for
each passenger class based on sex.  Use the cut method to create three bins to hold the passenger ages in
catagorical form.  Hint: use [0,18,25,99] and ['kid','young_adult','adult'].

```python
titanic_df = sn.load_dataset("titanic")
# Now we drop those rows where theres a NAN in Age
titanic_df = titanic_df.dropna(subset=["age"])
# We can also get rid of columns we are not using
titanic_df = titanic_df.drop(["parch", "sibsp"], axis=1)
titanic_df.head()
```

```python
titanic_df["Age_Range"] = pd.cut(titanic_df["age"], [0,18,25,99], labels=['kid','young_adult','adult'])
titanic_df["Age_Range"]
```

```python
titanic_df.groupby(["sex", "pclass"])["survived"].mean()
```
